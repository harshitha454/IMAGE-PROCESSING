```python
'''  a. Read a colour image, convert it into gray scale image.
 Increase and decrease the contrast and display all the three images and save. '''

import cv2
import matplotlib.pyplot as plt
import os

path="/content/sample_data/keyguard_default_wallpaper (1).webp"
directory="D:\MCA SEM2"

try:
  os.mkdir(directory)
except Exception as e:
  print(e)

img=cv2.imread(path)
gray=cv2.imread(path,cv2.IMREAD_GRAYSCALE)

bc=img*15
dc=img-15

plt.title("ORIGINAL IMAGE")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()

plt.title("GRAY IMAGE")
plt.imshow(cv2.cvtColor(gray,cv2.COLOR_BGR2RGB))
plt.show()

plt.title("INCREASED CONTRAST")
plt.imshow(cv2.cvtColor(bc,cv2.COLOR_BGR2RGB))
plt.show()

plt.title("DECREASED CONTRAST")
plt.imshow(cv2.cvtColor(dc,cv2.COLOR_BGR2RGB))
plt.show()

cv2.imwrite(directory+"gray.jpg",gray)
cv2.imwrite(directory+"bc.jpg",bc)
cv2.imwrite(directory+"dc.jpg",dc)

```

```python
''' b) Downscale the image by resizing with custom height and width 
while not preserving the aspect ratio. '''
import cv2
import matplotlib.pyplot as plt
import os

path="/content/sample_data/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path)

plt.title("ORIGINAL IMAGE")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()

down_h=300
down_w=200
down_p=(300,200)

down=cv2.resize(img,down_p,interpolation=cv2.INTER_LINEAR)

plt.title("DOWN_SCALED IMAGE")
plt.imshow(cv2.cvtColor(down,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
'''c) Scale the image vertically and horizontally and display all three images. '''

import cv2
import matplotlib.pyplot as plt
import os

path="/content/sample_data/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path)

plt.title("ORIGINAL IMAGE")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()

h=img.shape[0]
w=img.shape[1]

hor=cv2.resize(img,(300,h),interpolation=cv2.INTER_AREA)
ver=cv2.resize(img,(w,600),interpolation=cv2.INTER_AREA)


plt.title("HORIZONTALLY SCALED")
plt.imshow(cv2.cvtColor(hor,cv2.COLOR_BGR2RGB))
plt.show()

plt.title("VERTICALLY SCALED")
plt.imshow(cv2.cvtColor(ver,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
'''d) Downsampling and Upsampling an image while preserving the aspect ratio. '''
import cv2
import matplotlib.pyplot as plt
import os

path="/content/sample_data/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path)

plt.title("ORIGINAL IMAGE")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()

ds=cv2.resize(img,(0,0),fx=0.5,fy=0.5)
us=cv2.resize(img,(0,0),fx=2,fy=2)

plt.title("Down Sampling")
plt.imshow(cv2.cvtColor(ds,cv2.COLOR_BGR2RGB))
plt.show()

plt.title("Up Sampling")
plt.imshow(cv2.cvtColor(us,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
'''e) Resize the image by stretching, shrinking 
and zooming the same by using interpolation methods. '''

import cv2
import matplotlib.pyplot as plt
import os

path="/content/sample_data/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path)

plt.title("ORIGINAL IMAGE")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()

zoom=cv2.resize(img,(720,600),interpolation=cv2.INTER_AREA)
shrink=cv2.resize(img,(35,350),interpolation=cv2.INTER_NEAREST)
stretch=cv2.resize(img,(600,600),interpolation=cv2.INTER_LINEAR)

plt.title("ZOOMED IMAGE")
plt.imshow(cv2.cvtColor(zoom,cv2.COLOR_BGR2RGB))
plt.show()

plt.title("SHRINKED IMAGE")
plt.imshow(cv2.cvtColor(shrink,cv2.COLOR_BGR2RGB))
plt.show()

plt.title("STRETCHED IMAGE")
plt.imshow(cv2.cvtColor(stretch,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
'''f) Rotate the image to about 60 degree without scaling'''

import cv2
import matplotlib.pyplot as plt
import os

path="/content/sample_data/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path)

plt.title("ORIGINAL IMAGE")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()

rows,col=img.shape[:2]
M=cv2.getRotationMatrix2D((col/2,rows/2),60,1)
Rotated=cv2.warpAffine(img,M,(col,rows))

plt.title("Translated Image")
plt.imshow(cv2.cvtColor(Rotated,cv2.COLOR_BGR2RGB))
plt.show()
```
