```python
#'''B) Increase and decrease the brightness and Save it in a drive'''
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
path="/content/sample_data/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path)
directory="D:\MCA SEM2"
try:
  os.mkdir(directory)
except Exception as e:
  print(e)
intensity_matrix=np.ones(img.shape,dtype="uint8")*60
brt=cv2.add(img,intensity_matrix)
drk=cv2.subtract(img,intensity_matrix)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(brt,cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(drk,cv2.COLOR_BGR2RGB))
plt.show()
cv2.imwrite(directory+"brt.jpg",brt)
cv2.imwrite(directory+"drk.jpg",drk)
```

```python
''' E) Display the image such that the pixels having intensity values below
 than 50 will display as black and pixels having intensity values
  above than 150 will display as white. And the pixels between these will display as it is. 
'''
import cv2
import matplotlib.pyplot as plt
import numpy as np

path="/content/sample_data/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path,cv2.IMREAD_GRAYSCALE)
img=cv2.resize(img,(500,600))
plt.imshow(img)
plt.title("ORIGINAL IMAGE")
plt.show()

h=img.shape[0]
w=img.shape[1]

for i in range(0,h):
  for j in range(0,w):
    pix=img[i,j]
    if pix>150:
      img[i,j]=np.array([255])
    elif pix<50:
      img[i,j]=np.array([0])

print(pix)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Converted image")
plt.show()


```
