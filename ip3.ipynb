```python
''' 7) Write a python program for gray level slicing with or without background '''

#with background
import cv2
import matplotlib.pyplot as plt
import numpy as np

path="/content/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path,0)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()

r,c=img.shape[:2]
img1=np.zeros((r,c),dtype='uint8')
for i in range(r):
  for j in range(c):
    pix=img[i,j]
    if(pix>50 and pix<150):
       img1[i,j]=255
    else:
       img1[i,j]=pix

plt.imshow(img1)
plt.show()

#without background
r,c=img.shape[:2]
img1=np.zeros((r,c),dtype='uint8')
for i in range(r):
  for j in range(c):
    pix=img[i,j]
    if(pix>10 and pix<60):
       img1[i,j]=255
    else:
       img1[i,j]=0

plt.imshow(img1)
plt.show()

```

```python
'''9.	Write a python program for enhance the brightness 
and contrast using histogram equalization.'''

import cv2
import matplotlib.pyplot as plt
import numpy as np

path="/content/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path)

yuv=cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
yuv[:,:,0]=cv2.equalizeHist(yuv[:,:,0])
equalize=cv2.cvtColor(img,cv2.COLOR_YUV2BGR)

plt.imshow(img)
plt.show()
plt.imshow(yuv)
plt.show()
plt.imshow(equalize)
plt.show()


```

```python
#10.	Demonstrate edge detection of your image with first order.
# canny,sobel,perwitt,roberts

import cv2
import matplotlib.pyplot as plt
import numpy as np

path="/content/keyguard_default_wallpaper (1).webp"
img=cv2.imread(path)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()

#canny
canny=cv2.Canny(img,100,200)

#sobel
gaussia=cv2.GaussianBlur(img,[3,3],5)
sobel_x=cv2.Sobel(gaussia,cv2.CV_8U,1,0,ksize=5)
sobel_y=cv2.Sobel(gaussia,cv2.CV_8U,0,1,ksize=5)
sobel=sobel_x+sobel_y

#perwitt

kx=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
ky=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

px=cv2.filter2D(gaussia,-1,kernel=kx)
py=cv2.filter2D(gaussia,-1,kernel=ky)
perwitt=px+py
plt.imshow(cv2.cvtColor(gaussia,cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(canny,cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(sobel,cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(perwitt,cv2.COLOR_BGR2RGB))
plt.show()
```
